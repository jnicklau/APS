```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.ar_model import AutoReg, ar_select_order
from statsmodels.tsa.api import SARIMAX
from sklearn.metrics import mean_squared_error,r2_score
from math import sqrt
from matplotlib.markers import MarkerStyle
from matplotlib.lines import Line2D
```

```python
# Generate example time series data
np.random.seed(42)
n = 100
time = np.arange(n)
data = 0.5 * time + np.random.normal(scale=1, size=n)
```

```python
# Create a DataFrame
df = pd.DataFrame({'Time': time, 'Data': data})

# Split the data into training and testing sets
train_size = int(0.5 * len(df))
train, test = df[:train_size], df[train_size:]
```

```python
# Plot the time series data
plt.plot(df['Time'], df['Data'], label='Original Data')
plt.xlabel('Time')
plt.ylabel('Data')
plt.legend()
plt.show()
```

```python
# Fit an autoregressive model
order_of_ar_model = 10  
model = AutoReg(train['Data'], lags=order_of_ar_model,
                 trend = 'ct' 
               )
results = model.fit()
print(results.summary())
print(results.aic)
print((results.hqic))
print(results.aicc)
print(results.bic)
print(results.fpe)
print(results.df_model)
print(results.df_resid)
print(results.tvalues)
print(results.pvalues)
```

```python
sel = ar_select_order(train['Data'], maxlag = order_of_ar_model, ic = "aic", glob=True, old_names=False, trend = 'c')
sel.ar_lags
res_glob = sel.model.fit()
print(res_glob.summary())
print(res_glob.pvalues)
print(res_glob.tvalues)
```

```python
# Predict on the test set
predictions_res = results.predict(start=len(train), end=len(train) + len(test) - 1, dynamic=False)
predictions_res_glob = res_glob.predict(start=len(train), end=len(train) + len(test) - 1, dynamic=False)

# Evaluate the model
rmse = [sqrt(mean_squared_error(test['Data'], predictions_res)), sqrt(mean_squared_error(test['Data'], predictions_res_glob))]
r2score = r2_score(test['Data'], predictions_res),  r2_score(test['Data'], predictions_res_glob)
print('Root Mean Squared Error (RMSE): ', rmse)
print('R2-Score: ', r2score)
```

```python
# Plot the results
plt.plot(df['Time'], df['Data'], label='Original Data')
plt.plot(test['Time'], predictions_res, label='AR Model Predictions', linestyle='--', color='red')
plt.xlabel('Time')
plt.ylabel('Data')
plt.legend()
plt.show()
results.plot_diagnostics()
```

```python
fcast_instances = 10
start = train['Data'].index[-fcast_instances]
end = train['Data'].index[-1]
fcast_duration = 10
```

```python
fcast_index = list(range(start, end+fcast_duration+1))
print(fcast_index)
```

```python
forecasts = pd.DataFrame(index=fcast_index, columns=fcast_index)
gorecasts = pd.DataFrame(index=fcast_index, columns=fcast_index)
#print(forecasts)
```

```python
for i in range(0, fcast_instances):
    fcast = res_glob.predict(
        start=fcast_index[i], end=fcast_index[i + fcast_duration], dynamic=True
    )
    gcast = res_glob.predict(
        start=fcast_index[i], end=fcast_index[i + fcast_duration], dynamic=False
    )
    gorecasts.loc[fcast.index, fcast_index[i]] = gcast
    forecasts.loc[fcast.index, fcast_index[i]] = fcast
fig, ax = plt.subplots(figsize=(8, 5))
train['Data'].iloc[-fcast_instances:].plot(marker='o', linestyle='-', ax=ax, color="black")
test['Data'].iloc[:fcast_duration].plot(ax=ax, color="black", linestyle="-",marker='o')
ax = forecasts.plot(linestyle = '-',ax=ax,marker='.',fillstyle = Line2D.fillStyles[5])
ax = gorecasts.plot(linestyle = '-',ax=ax,marker='.',fillstyle = Line2D.fillStyles[5])
```

```python
sarimax_mod = SARIMAX(train['Data'], trend="c")
sarimax_res = sarimax_mod.fit()
predictions_sarimax_res = sarimax_res.predict(start=len(train), end=len(train) + len(test) - 1, dynamic=False)
```

```python
plt.plot(df['Time'], df['Data'], label='Original Data')
plt.plot(test['Time'], predictions_sarimax_res, label='AR Model Predictions', linestyle='--', color='red')
plt.xlabel('Time')
plt.ylabel('Data')
plt.legend()
plt.show()
```
