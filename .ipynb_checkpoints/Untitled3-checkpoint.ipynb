```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.tsa.ar_model import ar_select_order
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, r2_score
from sklearn import linear_model
```

```python
def smooth(a,n=1):
    adiff = np.diff(a)
    for i in range(n):
        a[:,1:] += adiff
    return a
def include_autodependence(a,n=1,c= 10):
    adiff = np.diff(a)
    for i in range(n):
        a[:,1:] *= adiff/c
    return a
def linear_regression_train(X, y, **kwargs):
    print("LinearRegression")
    lregr = linear_model.LinearRegression(positive=kwargs.get("positive", False))
    lregr.fit(X, y)
    return lregr
```

```python
b = np.random.rand(1,10)

plt.plot(b.reshape(-1,1))
plt.plot(smooth(b,10).reshape(-1,1),label = 'smooth')
plt.plot(include_autodependence(b,2).reshape(-1,1), label = 'include_autodependence')
plt.show()
```

```python
# Step 2: Generate simulated data
np.random.seed(10)
n_obs = 100
n_exog = 5
time = np.arange(n_obs)
exogenous_variable = np.zeros((n_exog,n_obs))
endogenous_variable = np.zeros((1, n_obs))
for i in range(n_exog):
    exogenous_variable[i,:] = np.random.randn(n_obs)  # Simulated exogenous variable
true_p = 2  # Order of autoregressive component in the true model
true_q = 1  # Order of moving average component in the true model
true_coefficients = np.array([0.5, -0.2, 0.8])  # Coefficients for autoregressive and moving average terms
error = np.random.normal(0, 1, n_obs)  # White noise error
endogenous_variable[:,:] = np.sum(1*exogenous_variable, axis=0)
print(np.shape(exogenous_variable))
print(np.shape(endogenous_variable))
#simulated_data = smooth(endogenous_variable,3) + error
#simulated_data = endogenous_variable + error
simulated_data = include_autodependence(endogenous_variable,3)/2 + error

print(np.shape(simulated_data))
```

```python
split = int(0.8 * n_obs)
```

```python
# Step 3: Fit SARIMAX model
order = (3, 0, 3)  # Example order for SARIMAX (p, d, q)
maxorder = 10
print(np.shape(simulated_data[:,:split].reshape(-1,1)))
print(np.shape(exogenous_variable[:,:split].reshape(-1,2)))
arimax = ARIMA( simulated_data[:,:split].reshape(-1,1), 
                    exog=exogenous_variable[0,:split].reshape(-1,1), 
                    order=order,
                   trend = 'n'
              )
arimax_result = arimax.fit()
print(arimax_result.summary())
```

```python
# Step 4: Simulate future values
n_forecast = n_obs-split
print(n_forecast)
print(np.shape(exogenous_variable[0,-n_forecast:]))
forecast = arimax_result.get_forecast(steps=n_forecast, exog=exogenous_variable[0,-n_forecast:])
forecast_index = np.arange(split, n_obs)
```

```python
# Step 5: Simulate linear regression
model = linear_regression_train(exogenous_variable[0,:split].reshape(-1,1), simulated_data[0,:split].reshape(-1,1))
prediction_lreg = model.predict(exogenous_variable[0,split:].reshape(-1,1))
```

```python
# Calculate R squared on the  forecast
r2 = r2_score(simulated_data[:,split:].ravel(), forecast.predicted_mean)
print(r2)
# Calculate R squared on the lregr_prediction
r2 = r2_score(simulated_data[:,split:].ravel(), prediction_lreg.ravel())
print(r2)
# Calculate average deviation
print(np.sqrt(np.sum(prediction_lreg-forecast.predicted_mean)**2)/n_forecast)
```

```python
# Step 6: Simulate future values using linear regressed values
forecast_lregr = arimax_result.get_forecast(steps=n_forecast, exog= prediction_lreg)
print(forecast_lregr)
```

```python
arimax_result.plot_diagnostics()
```

```python
# Plot the simulated data and forecast
print(np.shape(forecast_lregr))

plt.plot(forecast_index, simulated_data[:,split:].ravel(), label='Simulated Data',color = 'black')
plt.plot(forecast_index, forecast.predicted_mean,'o', label='Forecast', linestyle='--', color='red')
plt.fill_between(forecast_index, forecast.conf_int()[:, 0], forecast.conf_int()[:, 1], color='red', alpha=0.2)
plt.plot(forecast_index, prediction_lreg,'o', label='LReg', linestyle='--', color='blue')
plt.plot(forecast_index, forecast_lregr.predicted_mean,'o', label='LReg + Forecast', linestyle='-', color='purple')

plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.show()
```
